```python
from geomstats.geometry.hypersphere import Hypersphere
from geomstats.learning.principal_nested_spheres import PrincipalNestedSpheres
from geomstats.visualization.hypersphere import Sphere, Circle
```

```python
sphere = Hypersphere(dim=2)
X = sphere.random_uniform(n_samples=100)
```

```python
pns = PrincipalNestedSpheres(sphere=sphere)
pns.fit(X)
X_reduced = pns.transform(X)
```

```python
X_reduced
```

```python
plot_X = Sphere(points=X)
plot_X.plot(X)
```

```python
plot_X_reduced = Circle(points=X_reduced)
plot_X_reduced.plot(X_reduced)
```
