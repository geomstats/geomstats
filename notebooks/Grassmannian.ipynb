# I. Introduction to the Grassmannian

The collection of all linear subspaces with a dimension V within the Euclidean space $\mathbb{R}^k$ form the Grassman manifold, Gr(n,k). The Grassmanian can be expressed in various ways: a set, a differentiable manifold, a set of orthogonal projections, a homogeneous space, or a scheme.  One of the most significant qualities of a Grassmannian manifold is that they are an orbit space of the Steifel Manifold $\mathcal{v_{n,k}}$ of orthonormal $k$ frames in $\mathbb{G}^n$. (Wolfram)

# II.  Mathematical Definition of the Grassmannian 

The Grassmannian Gr(n,k) is the set of k-dimensional subspaces in an n-dimensional vector space. Both real and complex Grassmannians are manifolds. These provide a setting for comparing distances between subspaces using a variety of metrics. (Turaga et al. 2008)

The geometric definition of the Grassmanian manifold is as follows.

Let *V* be the *n* dimmensional vector space over a field *K*. The Grassmanian (**Gr**(*k*,*V*)) is the set of all k-dimmensional linear subspaces in *V*. As a mathmatical definition, this can be expressed as:

\begin{align}
    Gr(k,V): { \mathcal{U} \subset \mathbb{R^k} \mid \mathcal{U}\ is\ a\ subspace,\ dim\ (\mathcal{U}) = V }
    \end{align}

For the case where $k$=1 and $V$=2 the Grassmannian manifold reduces to the real projective space which consists of all lines through the origin. 



# III. Real World Applications of the Grassmannian Manifold

**Grassmanian Manifolds used to classify hyperspectral image data  of the indian Pine Test site** Ma et al. (2019)

The Structure of the Grassmannian is used to measure the distance between two points, and a method to move one point in the direction of the other. This means that the data does not need to have a representation in Euclidean space. In this algorithm, a point on the Grassmannian is moved in the direction of another along a geodesic path. This is implemented on several real life data sets such as hyperspectral images of the indian Pine Test sites. In the example below, different areas of the Indian Pine test are classified into classes such as corn(green), corn-notill(red),corn-mintill( blue).

pine_grassmannian_3.png

**Grassmannian manifolds for metastatic liver segmentation** Kadoury et al. (2015)

A current real world application of the Grassmannian manifold is its use in the segmentation of biomedical images. In this specific example, a Grassmanian manifold is used to extract tumors from medical images of livers. It involves a training phase where the Grassmanian manifold is trained using tumor and healthy liver tissue images that were extracted manually from a data set of sample images. In the testing phase the patches extracted from the images are projected onto a Grassmanian manifold.


tumor_grassmannian_2.jpg

**Grassmannian manifold for Alzheimers disease** Gui et al. (2017)

Sequential MRI brain scans were taken over time of individuals with either a healthy aging background, Alzheimer's disease, or mild cognitive impairment. The shape of 6 structures of interest - the bilateral hippocampi, amygdalas, and lateral ventricle - was evaluated by calculating the geodesic distance on the Grassmanian manifold. Statistical quantification shows that changes in geodesic distance link with changes in cognitive measures (AD>MCI>HC).


alzheimers_grassmanian.png

**Grassmanian manifolds in video-based human emotion recognition** Liu et al. (2013)

Model all video frames of a persons face as an emotional image set. The set of images can be modeled by a linear subspace which characterizes the specific persons emotion. For distance metric, the collection of subspaces are treated as points on Grassmannian manifold, and point-to point distances are induced from the Grassmannian kernels.
