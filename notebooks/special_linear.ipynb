## Checking updated special_linear.py file under geomstats.geomtery
Checked all functions in the file, they all work now

```python
from geomstats.geometry.special_linear import *
import numpy as np
```

```python
instance = SpecialLinear(n=2)
print(instance)
```

```python
random1 = instance.random_point()
random2 = instance.random_point()
print(random1,random2)
```

```python
print(instance.belongs(random1))
print(instance.belongs(random2))
```

```python
my_matrix = np.array([[1, 1], [0, 0.7]])
my_point = instance.projection(my_matrix)
print(instance.belongs(my_point))
```

```python
algebra = TracelessMatrices(2)
print(algebra)
```

```python
random_alg1 = algebra.random_point()
random_alg2 = algebra.random_point()
print(random_alg1, random_alg2)
```

```python
print(algebra.belongs(random_alg1))
print(algebra.belongs(random_alg2))
```

```python
my_algebra = np.array([[1, -1], [0, -0.7]])
my_point_alg = algebra.projection(my_algebra)
print(algebra.belongs(my_point_alg))
```

```python
basis1 = np.array([[1,0],[0,1]])
basis2 = np.array([[0,1],[-1,0.2]])
representation1 = algebra.basis_representation(basis1)
representation2 = algebra.basis_representation(basis2)
print(representation1,representation2)
```
